### Using loc for proper allocation and functions

```python
# setting values like previous method leads to confusion at the time of returning so allocation is done by:
newdf.loc[0,0] = 654
```

```python
newdf.head(2)
```

```python
newdf.columns = list("ABCDEF")
```

```python
newdf
```

```python
newdf.loc[0, 'A']= 7
```

```python
newdf.head(2)
```

```python
newdf= newdf.drop('F', axis=1)    #for deleting a row or a column. By default the axis=0
```

```python
newdf.head()
```

```python
#If we want to extract specific rows or columns
newdf.loc[[0,4], ['B', 'E']]  #for all the rows or columns we can use colon
```

```python
newdf.loc[(newdf['A']<0.3) & (newdf["E"]>0.2)]  #& is used in place of comma for complex query
```

```python
newdf.iloc[0,4]  #this will print the value of 0,4 INDEX. loc for name, iloc for index
```

```python
newdf.iloc[[0,4], [4,2]]   
```

**All these functions we are using do not affect the original newdf as we are not allocating newdf= function

**NaN is knows as not a number and dropna function can be used to remove them**

```python
dicti = {'First Score':[100, 90, np.nan, 95],
        'Second Score': [np.nan, np.nan, np.nan, np.nan],
        'Third Score':[np.nan, 40, 80, 98]}
nf=pd.DataFrame(dicti)
```

```python
nf
```

```python
nf.dropna(how='all' , axis=1, inplace=True)
```

```python
nf
```

**drop_duplicates is used to remove duplicates and it takes subset argument instead of axis**

```python
g = {'name':["aparna", "pankaj", "sudhir", "Geeku"],
        'degree': ["MBA", "MBA", "M.Tech", "MBA"],
        'score':[90, 40, 80, 98]}
df1= pd.DataFrame(g)
```

```python
df1
```

```python
df1.drop_duplicates(subset=['degree'])
```

```python
df1.drop_duplicates(subset=['degree'], keep='first')
```

```python
df1.drop_duplicates(subset=['degree'], keep='last')
```

```python
df1.drop_duplicates(subset=['degree'], keep= False)
```

```python
df1.shape
```

```python
df1.info()
```

```python
df1['degree'].value_counts(dropna=True)
```
