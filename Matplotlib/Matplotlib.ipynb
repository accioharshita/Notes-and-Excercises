```python
from collections import Counter
import pandas as pd

data= pd.read_csv('data.csv')

ids= data['Responder_id']
lang= data['LanguagesWorkedWith']

c= Counter()

for i in lang:
    c.update(i.split(';'))

languages=[]
popularity=[]

for item in c.most_common():                  #this returns information in a list
    languages.append(item[0])
    popularity.append(item[1])

print(languages)
print(popularity)
```

```python
data= pd.read_csv('data.csv')

ids= data['Responder_id']
lang= data['LanguagesWorkedWith']

c= Counter()

for i in lang:
    c.update(i.split(';'))

languages=[]
popularity=[]

for item in c.most_common():                  
    popularity.append(item[1])
    languages.append(item[0])


plt.barh(languages, popularity)


plt.title('Popularity of Programming languages')
plt.xlabel('Popularity')

plt.show()
```

## Filling area on line plots

```python
age= d['Age']
dev= d['All_Devs']
python=d['Python']
js= d['JavaScript']

plt.plot(age,dev, label='All Dev')
plt.plot(age,python, label='Python Dev')

#median=57287
plt.fill_between(age, python, alpha= 0.25) #alpha is the parameter for sheerness, default set to 0(solid fill)

#fill between takes two values for plotting, for more than two there's a value called y2=0 (default)

plt.xlabel('Ages')
plt.ylabel('Median Salary(USD)')
plt.legend()
plt.show()
```

**with a conditional thershold, the y2 parameter:**

```python
age= d['Age']
dev= d['All_Devs']
python=d['Python']
js= d['JavaScript']

plt.plot(age,dev, label='All Dev')
plt.plot(age,python, label='Python Dev')

median=57287
plt.fill_between(age, python, median, where=(python>median), interpolate=True, alpha= 0.25)

#where parameter sets a condition for customising the fill, like the colour,opacity etc.
#the interpolate prevents any kinda chaffing or loss of data on x axis

plt.fill_between(age, python, median, where=(python<=median), interpolate=True, alpha= 0.25)

#having two fill statements for two different conditions, produces different fills. We can customise it with colors too.


plt.xlabel('Ages')
plt.ylabel('Median Salary(USD)')
plt.legend()
plt.show()

#It clearly shows that when the plot is below median salary, it fills up else, fills down to it.
```

```python
age= d['Age']
dev= d['All_Devs']
python=d['Python']
js= d['JavaScript']

plt.plot(age,dev, label='All Dev')
plt.plot(age,python, label='Python Dev')


plt.fill_between(age, python, js, where=(python>js), interpolate=True, alpha= 0.25, label= 'Above Avg')


plt.fill_between(age, python, js, where=(python<=js), interpolate=True, alpha= 0.25, label= 'Below Avg')

#It's a good idea to label the fills as it will give a clear understanding
#We can do that by passing the labels parameter in the fill arg


plt.xlabel('Ages')
plt.ylabel('Median Salary(USD)')
plt.legend()
plt.show()
```

## Plotting Time-Series Data

```python
from datetime import datetime, timedelta
from matplotlib import dates as mpl_dates

dates = [ datetime(2019, 5, 24),
    datetime(2019, 5, 25),
    datetime(2019, 5, 26),
    datetime(2019, 5, 27),
    datetime(2019, 5, 28),
    datetime(2019, 5, 29),
    datetime(2019, 5, 30)
        ]

y = [0, 1, 3, 4, 6, 5, 7]

plt.plot_date(dates, y, linestyle='solid')


#if you want to change the positon of the dates then we use an autoformat function:
'''plt.gcf().autofmt_xdate() (gcf= get current figure)
which rotates the dates on x axis'''

#now if you want the format of yyyymmdd of the date:
'''Import dates from mpl and then the below code.
%d= date,   %b= month,  %Y= Year'''

date_format= mpl_dates.DateFormatter('%d %b, %Y')
plt.gca().xaxis.set_major_formatter(date_format)
#gca means get current axes


plt.tight_layout()
plt.show()
```

**Plotting Live Data in Real-Time**


```python
%matplotlib notebook    
 
#This makes your live graphs appear on matplotlib and all graphs with more quality and control

from matplotlib.animation import FuncAnimation
import random
from itertools import count
import matplotlib.pyplot as plt


x=[]
y=[]

index= count()



def animate(i):
    x.append(next(index))     #used to give the next value from the iterator
    y.append(random.randint(0,5))
    plt.cla()       #to clear the axis so that a single coloured line is obtained instead of different colours
    plt.plot(x,y)

    
ani= FuncAnimation(plt.gcf(), animate, interval=1000)  #here 1000= 1sec

plt.tight_layout()
plt.show()
```
